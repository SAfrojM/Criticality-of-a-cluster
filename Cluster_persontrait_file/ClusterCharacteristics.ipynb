```python
shapefile1["NEIGHBORS"] = None  
adjList = pd.DataFrame(columns=['source', 'target'])

zipcodetoIndex=dict(zip(shapefile1.GEOID10, np.arange(shapefile1.GEOID10.shape[0])))
indextoZipcode=dict(zip(np.arange(shapefile1.GEOID10.shape[0]),shapefile1.GEOID10))
```

```python
nodePropfromPerTrait= pd.read_csv("/home/cbn7xr/Measles/nodePropfromPerTrait.csv")
nodePropfromPerTrait
```

```python
clus=cluster3
pop=0
nVacc=0
land=0
for c in clus:
    print(indextoZipcode[c])
    pop=pop+(nodePropfromPerTrait.loc[nodePropfromPerTrait['zipcode']==int(indextoZipcode[c])]['population'].sum())
    nVacc=nVacc+(nodePropfromPerTrait.loc[nodePropfromPerTrait['zipcode']==int(indextoZipcode[c])]['nVaccCount'].sum())
    land=land+shapefile1.loc[shapefile1['ZCTA5CE10']==(indextoZipcode[c])]['ALAND10'].sum()
    
print('pop='+str(pop))   
print('nVacc='+str(nVacc))
print('nVaccPer='+str(nVacc*100/pop))
print('land='+str(land*.0000003861))
print('density='+str(pop/(land*.0000003861)))
```
